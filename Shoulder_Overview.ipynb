```python
import matplotlib.pyplot as plt
import numpy as np
import getsig as getsig
from rpsavgperf import rpsavgperf
from perf2rho import perf2rho
plt.style.use('helvet2')
```

```python
shots_performed = [34102, 34103, 34104, 34105, 34106, 34107, 34108, 34275, 34276, 34277, 34278, 34279, 34280, 34281]
```

```python
shotnr = 34281
```

```python
times = np.array([1.9, 3.0, 5.0, 6.0])
l0 = rpsavgperf(shotnr, time=times[0])
h0 = rpsavgperf(shotnr, time=times[0], side=0)
l1 = rpsavgperf(shotnr, time=times[1])
h1 = rpsavgperf(shotnr, time=times[1], side=0)
l2 = rpsavgperf(shotnr, time=times[2])
h2 = rpsavgperf(shotnr, time=times[2], side=0)
l3 = rpsavgperf(shotnr, time=times[3])
h3 = rpsavgperf(shotnr, time=times[3], side=0)
```

```python
rl0 = perf2rho(shotnr, times[0], l0.r, 0.14)
rl1 = perf2rho(shotnr, times[1], l1.r, 0.14)
rl2 = perf2rho(shotnr, times[2], l2.r, 0.14)
rl3 = perf2rho(shotnr, times[3], l3.r, 0.14)
```

```python
plt.figure(dpi=150)
plt.semilogy(rl0,l0.n*1e-19, label='t=1.9s')
plt.semilogy(rl1,l1.n*1e-19, label='t=3.0s')
plt.semilogy(rl2,l2.n*1e-19, label='t=5.0s')
plt.semilogy(rl3,l3.n*1e-19, label='t=6.0s')
plt.xlim(0.9,1.15)
plt.title(r'$\mathrm{\#34281}$', loc='left')
plt.xlabel(r'$\mathrm{\rho_{pol}}$')
plt.ylim(0,2.8)
plt.ylabel(r'$\mathrm{n_{e}\,[10^{19}m^{-3}]}$')
plt.axvline(1.0, ls='--', color='k')
plt.axvspan(0.9,1.0, color='k', alpha=0.2)
plt.legend()
plt.tight_layout()
plt.savefig('log34281.png',dpi=300)
plt.show()
```
